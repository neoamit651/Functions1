6. Explain the concept of generators in Python and how they are defined.
  - A generator in Python is a special type of iterator that allows you to create a sequence of values on the fly, without storing them all in memory at once. This makes generators highly memory-efficient and suitable for processing large or even infinite datasets.
Unlike a regular function that returns a single value and then terminates, a generator function uses the yield keyword to return a value and temporarily suspend its execution. The next time a value is requested, the generator resumes from where it left off, maintaining its state.
  - How generators work
    - The magic of a generator lies in the yield keyword:
      -1. Pauses execution: When yield is encountered, the function's state is frozen, including the values of all local variables. The yielded value is returned to the caller.
      -2. Resumes execution: When the next value is requested (e.g., in a for loop or with the next() function), the generator's execution resumes from the exact point it was paused.
      -3. Raises StopIteration: Once the generator has yielded all of its values and finishes, subsequent calls to next() automatically raise a StopIteration exception. A for loop handles this exception gracefully and terminates
  - How to define generators
    - Generators can be defined in two ways: using a generator function or a generator expression.
    - 1. Generator function
      - A function becomes a generator function if it contains at least one yield statement.
      - Example: A generator that yields an infinite sequence of even numbers.
        - def even_numbers():
          - n = 0
          - while True:
              - yield n
              - n += 2

        -"""Create the generator object"""
        - even_gen = even_numbers()
        -""" Get the first few values"""
        - print(next(even_gen)) # Output: 0
        - print(next(even_gen)) # Output: 2
        - print(next(even_gen)) # Output: 4
      
      - Example: A generator that yields a finite sequence.
        - def count_up_to(max_value):
        - current = 1
        - while current <= max_value:
          - yield current
        - current += 1

        - for number in count_up_to(5):
            - print(number)
        - Output:
        -  1
        -  2
        -  3
        -  4
        -  5  
      - 2. Generator expression
        - Similar to list comprehensions, generator expressions provide a concise way to create generators. They use parentheses () instead of the square brackets [] used by list comprehensions.
        - Example: A generator expression to get the squares of numbers from 1 to 5.
        - squares_gen = (x**2 for x in range(1, 6))
        - """ The expression returns a generator object, not a list"""
        - print(squares_gen)
        
        - Output: <generator object <genexpr> at ...>

        - """Iterate over the generator to get the values"""
        - for square in squares_gen:
          - print(square)
        -  Output:
        -  1
        -  4

7. What are the advantages of using generators over regular functions?
  - Comparison at a glance
  
  123.jpg

  - Generators offer significant advantages over regular functions, particularly in terms of memory efficiency, processing large or infinite data streams, and controlling execution flow. While a regular function computes and returns all results at once, a generator uses the yield keyword to produce values one at a time, pausing and resuming its execution as needed.
    - Key advantages of using generators
      - Memory efficiency : For functions that need to return a large sequence of results, a regular function builds the entire sequence in memory before returning it. A generator, however, yields one item at a time, making it highly memory-efficient, especially for big data.
          - Regular function: Returns a full list or array, which can lead to memory errors if the dataset is too large to fit in memory.
          - Generator function: Processes data lazily, generating values on-the-fly. This results in a much smaller, constant memory footprint, regardless of the sequence size.
      - Lazy evaluation: Generators calculate and produce values only when they are explicitly requested. This "lazy" or "on-demand" evaluation saves computational resources and can improve performance by avoiding unnecessary work.
          - Regular function: Eagerly evaluates all results upfront, even if the caller only needs the first few.
          - Generator function: Defers computation until the next() method is called on the generator object.
      - Handling infinite sequences
        - Generators can easily represent infinite data streams, which is impossible for a regular function that must return a finite collection.
          - Regular function: Cannot be used for infinite sequences because it would never finish building the data structure to return.
          - Generator function: Can run in an infinite loop, yielding the next item in the sequence each time it is called, without ever consuming excessive memory.
      - Simplified code - Generators are simpler and cleaner to write than manually building an iterator class, which can involve defining several special methods. The logic is encapsulated concisely within a single function.
      - Execution control : A generator function can be paused and resumed, allowing for more complex and flexible control flows.
        - Regular function: Runs from start to finish in a single pass, and its local state is destroyed upon returning.
        - Generator function: Preserves its internal stateincluding local variablesbetween yield calls. When resumed, it continues from where it left off.

9. Explain the purpose and usage of the `map()` function in Python.
  - The map() function in Python applies a specified function to each item in an iterable and returns an iterator of the results. It provides a concise, functional-programming approach to perform uniform transformations on data without needing an explicit for loop.
    - Purpose
      - The primary purpose of map() is to transform every element of a sequence in the same way. It is especially useful for:
      - Concise code: You can replace a multi-line for loop with a single, readable line of code.
      - Memory efficiency: In Python 3, map() returns a lazy iterator (a map object). This means it processes elements one at a time and does not create a new, full list in memory until it's needed. This is highly efficient for very large datasets.
      - Multiple iterables: It can apply a function to multiple iterables in parallel, as long as the function takes a matching number of arguments.
  - Usage and syntax
    - The syntax for the map() function is:
    - map(function, iterable, [iterable2, ...])
        - function: The callable to be applied to each item. This can be a built-in function, a user-defined function, or a lambda function.
        - iterable: The sequence (e.g., list, tuple, string) of items to be processed.
        - [iterable2, ...]: Optional additional iterables.
  -The return value is a map object, which is an iterator. To see the results, you typically convert it to a list, tuple, or other sequence type.
  - Using map() with a lambda function
      - numbers = [1, 2, 3, 4]
      - doubled_numbers = map(lambda x: x * 2, numbers)
      - print(list(doubled_numbers))
      
      - Output: [2, 4, 6, 8]

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
# Ans.
import math

def is_prime(num):
    """
    Checks if a given number is prime.

    Args:
        num: An integer to be checked for primality.

    Returns:
        True if the number is prime, False otherwise.
    """
    if num <= 1:
        return False  # Numbers less than or equal to 1 are not prime
    if num == 2:
        return True   # 2 is the only even prime number
    if num % 2 == 0:
        return False  # Other even numbers are not prime

    # Check for divisibility by odd numbers from 3 up to the square root of num
    # We only need to check up to the square root because if a number has a
    # divisor greater than its square root, it must also have a divisor
    # smaller than its square root.
    for i in range(3, int(math.sqrt(num)) + 1, 2):
        if num % i == 0:
            return False  # Found a divisor, so it's not prime
    return True  # No divisors found, so it's prime

def find_primes_in_range(start, end):
    """
    Finds all prime numbers within a specified range.

    Args:
        start: The beginning of the range (inclusive).
        end: The end of the range (inclusive).

    Returns:
        A list containing all prime numbers in the given range.
    """
    prime_numbers = []
    for number in range(start, end + 1):
        if is_prime(number):
            prime_numbers.append(number)
    return prime_numbers

# Example usage for numbers from 1 to 200
primes_up_to_200 = find_primes_in_range(1, 200)
print(f"Prime numbers from 1 to 200: {primes_up_to_200}")
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
# Ans.
def powers_of_2(exponent):
    """
    Yields the powers of 2 up to a given positive or negative exponent.

    Args:
        exponent: An integer representing the maximum power of 2 to yield.

    Yields:
        int or float: The next power of 2.
    """
    if not isinstance(exponent, int):
        raise TypeError("The exponent must be an integer.")

    if exponent >= 0:
        for i in range(exponent + 1):
            yield 2**i
    else:
        # For negative exponents, yield from 0 down to the given exponent.
        for i in range(0, exponent - 1, -1):
            yield 2**i

# Example usage with a positive exponent
print("Powers of 2 up to exponent 5:")
for p in powers_of_2(5):
    print(p)

# Example usage with a negative exponent
print("\nPowers of 2 down to exponent -3:")
for p in powers_of_2(-3):
    print(p)

# Example with exponent 0, which yields 2**0 = 1
print("\nPowers of 2 up to exponent 0:")
for p in powers_of_2(0):
    print(p)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
# Ans.
import os

def read_file_lines(file_path):
    """
    A generator function that reads a file line by line.

    Args:
        file_path (str): The path to the file to be read.

    Yields:
        str: Each line of the file, with trailing whitespace stripped.
    """
    if not os.path.isfile(file_path):
        raise FileNotFoundError(f"The file at '{file_path}' was not found.")

    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            yield line.rstrip('\n')

# --- Example Usage ---

# 1. First, create a dummy file for testing.
file_name = 'sample.txt'
with open(file_name, 'w', encoding='utf-8') as f:
    f.write("First line of text.\n")
    f.write("Second line with some content.\n")
    f.write("Third and final line.\n")

# 2. Use the generator function to read the file line by line.
print("Reading lines from 'sample.txt' using the generator:")
try:
    for line in read_file_lines(file_name):
        print(line)
except FileNotFoundError as e:
    print(e)

# 3. Clean up the created file.
os.remove(file_name)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# Ans.
# A list of temperatures in Celsius
celsius_temps = [0, 10, 20, 30, 37, 100]

# Use a lambda function with map() to convert each temperature to Fahrenheit
# The lambda function takes a single argument 'c' and applies the conversion formula
fahrenheit_temps = map(lambda c: (c * 9/5) + 32, celsius_temps)

# The map() function returns a map object, so we convert it to a list
# to print the results
fahrenheit_list = list(fahrenheit_temps)

# Print the list of converted temperatures
print("Celsius temperatures:", celsius_temps)
print("Fahrenheit temperatures:", fahrenheit_list)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
# Ans.
def remove_vowels(input_string):
    """
    Removes all vowels (uppercase and lowercase) from a string using filter().

    Args:
        input_string: The string to process.

    Returns:
        A new string with all vowels removed.
    """
    vowels = 'aeiouAEIOU'

    # Define a lambda function that returns True if a character is not a vowel.
    # The filter() function will then keep only the characters for which
    # the lambda function returns True.
    filtered_chars = filter(lambda char: char not in vowels, input_string)

    # The filter() function returns an iterator, so we use ''.join()
    # to combine the filtered characters into a new string.
    result_string = ''.join(filtered_chars)
    return result_string

# Example usage
original_text = "Hello, Python is a great language!"
vowel_free_text = remove_vowels(original_text)

print(f"Original string: {original_text}")
print(f"String with vowels removed: {vowel_free_text}")

# Example with a different string
another_text = "Programming is fun"
vowel_free_another_text = remove_vowels(another_text)

print(f"\nOriginal string: {another_text}")
print(f"String with vowels removed: {vowel_free_another_text}")
```

```python
# 11) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this: Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,-  if the value of the order is smaller than 100,00 . Write a Python program using lambda and map.
# Ans.
# The list of book orders, with each sublist containing:
# [Order Number, Book Title and Author, Quantity, Price per Item]
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfhrung in Python3, Bernd Klein", 3, 24.99]
]

# Use map() with a lambda function to calculate the total for each order.
# The lambda function processes each sublist (order) as follows:
# 1. Calculates the base price: order[2] * order[3] (Quantity * Price).
# 2. Applies the conditional logic:
#    - If the base price is less than 100, add 10 to it.
#    - Otherwise, add 0.
# 3. Returns a 2-tuple: (order[0], final_price)
result_tuples = list(map(
    lambda order: (order[0], order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0)),
    orders
))

# Print the final list of 2-tuples
print(result_tuples)
```
